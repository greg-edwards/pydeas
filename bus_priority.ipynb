```python
# import bus stop data from Bay of Plenty
import osmnx as ox

tags = {'highway':'tertiary'}
gdf = ox.geometries_from_place("Tauranga City, New Zealand", tags=tags)

gdf.head()
```

```python
import leafmap

m = leafmap.Map(center=[20, 0], zoom=3, widescreen=True)
m.add_gdf(gdf, layer_name="New Plymouth Bus Routes")
m
```
