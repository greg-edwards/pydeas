```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from datetime import time
```

```python
df = pd.read_csv(r'C:\Users\GSE1\Beca\3813283 - PT Services and Infrastructure SSBC - Documents\Technical - Working Files\01 SSBC\05 Shortlist\Headways\headways.csv')
df['Headway'] = df['Headway'].astype('int')
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time
df = df.pivot_table(index='Route Name', columns='Time', values='Headway')
```

```python
heat_map = sns.heatmap(df, cmap='RdBu', cbar_kws={'label': 'Headway (mins)'}, linewidths=.5)
plt.savefig('myfile.png', bbox_inches="tight", dpi=300)
```
