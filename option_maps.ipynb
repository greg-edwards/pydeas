```python
import pandas as pd
import geopandas as gpd
import numpy as np
import folium
```

# Option 1

```python
# data

intersections = gpd.read_file(r'C:\Users\GSE1\Beca\3813283 - PT Services and Infrastructure SSBC - Documents\Technical - Working Files\01 SSBC\05 Shortlist\Modelling specification\Op1\intersections_op1.geojson')
links = gpd.read_file(r'C:\Users\GSE1\Beca\3813283 - PT Services and Infrastructure SSBC - Documents\Technical - Working Files\01 SSBC\05 Shortlist\Modelling specification\Op1\links_op1.geojson')
```

```python
# map

option_1 = links.explore(
    tiles="CartoDB positron",
    legend=False,
    zoom_on_click=True,
    highlight=True,
    tooltip=False,
    popup=False,
    name="Midblocks - Option 1",
        style_kwds= {
        "color":"red",
        "weight":6,
        "opacity":0.5})

intersections.explore(
    m=option_1,
    legend=False,
    tooltip=False,
    popup=False,
    zoom_on_click=True,
    highlight=True,
    name="Intersections - Option 1",
    marker_type="circle_marker",
    marker_kwds={
        "radius" : "4"},
    style_kwds= {
        "fillColor":"yellow",
        "stroke":True,
        "weight":0.6,
        "opacity":0.7})

folium.LayerControl().add_to(option_1)
   
option_1
```

```python
option_1.save("option_1.html")
```
