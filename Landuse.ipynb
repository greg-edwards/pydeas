```python
import osmnx as ox

#resi land use in New Plymouth
tags = {'landuse' : 'residential'}
gdf_resi = ox.geometries_from_point((-39.057882, 174.073968), tags=tags, dist=25000)

gdf_resi.head()
```

```python
gdf_resi.plot(column='landuse', legend=True)
```

```python
#industrial and commerical land use in New Plymouth
tags = {'landuse' : ['industrial','commercial']}
gdf_work = ox.geometries_from_point((-39.057882, 174.073968), tags=tags, dist=25000)

gdf_work.head()
```

```python
#add resi landuse and bus routes to map

import leafmap.foliumap as leafmap
import leafmap.colormaps as cm

m = leafmap.Map()
m.add_data(gdf_resi, column='landuse', layer_name="Residential", zoom_to_layer=True, colors='blue', add_legend=True)

m.add_data(gdf_work, column='landuse', layer_name="Commercial/Industrial", zoom_to_layer=True, colors='green')
m.add_data(data=r'C:\Users\GSE1\Beca\3823121 - New Plymouth Bus Network Review - Documents\Job Delivery\Technical - Working Files\GIS\Bus_routes_2022.geojson', column='Name', layer_name='Bus Routes', legend_title='Bus Routes', cmap='viridis')

m
```

```python
m.to_html("Taranaki_landuse.html")
```
