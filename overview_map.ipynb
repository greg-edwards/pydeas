```python
import pandas as pd
import geopandas as gpd
import numpy as np
import folium
```

```python
#data

links_key = gpd.read_file(r'C:\Users\GSE1\Beca\3813283 - PT Services and Infrastructure SSBC - Documents\Technical - Working Files\01 SSBC\05 Shortlist\02 Infrastructure Options assessment\Delay analysis\Overview map\links_key_locations.geojson')
intersections_key = gpd.read_file(r'C:\Users\GSE1\Beca\3813283 - PT Services and Infrastructure SSBC - Documents\Technical - Working Files\01 SSBC\05 Shortlist\02 Infrastructure Options assessment\Delay analysis\Overview map\intersections_key_locations.geojson')
links_outside = gpd.read_file(r'C:\Users\GSE1\Beca\3813283 - PT Services and Infrastructure SSBC - Documents\Technical - Working Files\01 SSBC\05 Shortlist\02 Infrastructure Options assessment\Delay analysis\Overview map\links_outside_primary.geojson')
intersections_outside = gpd.read_file(r'C:\Users\GSE1\Beca\3813283 - PT Services and Infrastructure SSBC - Documents\Technical - Working Files\01 SSBC\05 Shortlist\02 Infrastructure Options assessment\Delay analysis\Overview map\intersections_outside_primary.geojson')
routes = gpd.read_file(r'C:\Users\GSE1\Beca\3813283 - PT Services and Infrastructure SSBC - Documents\Technical - Working Files\01 SSBC\05 Shortlist\01 Service options assessment\hybrid_services.geojson')
```

```python
#remove duplicates

links_key.drop_duplicates(subset='geometry', inplace=True, keep='first')
intersections_key.drop_duplicates(subset='geometry', inplace=True, keep='first')
links_outside.drop_duplicates(subset='geometry', inplace=True, keep='first')
intersections_outside.drop_duplicates(subset='geometry', inplace=True, keep='first')
```

```python
links_key.columns
```

```python
map = routes.explore(tiles="cartoDB positron",
    legend=True,
    column='line_name',
    zoom_on_click=True,
    tooltip=['line_name', 'hours_year', 'cost_year', 'vehicle_co', 'minheadway', 'maxheadway'],
    highlight=True,
    name="Hybrid Routes",
    popup=['line_name', 'hours_year', 'cost_year', 'vehicle_co', 'minheadway', 'maxheadway'],
    style_kwds= {
        "weight":3,
        "opacity":0.6,
        "dashArray": '5, 10'},
    legend_kwds={
        "caption": "Line Name"
    })

links_key.explore(
    m=map,
    legend=False,
    zoom_on_click=True,
    tooltip=['ADT_PT', 'AM_PT', 'LOS'],
    highlight=True,
    name="Links - Priority Locations",
    popup=['ADT_PT', 'AM_PT', 'LOS'],
    style_kwds= {
        "color":"red",
        "weight":6,
        "opacity":0.7})

intersections_key.explore(
    m=map,
    legend=False,
    zoom_on_click=True,
    tooltip=['LABEL', 'ADT_PT', 'AM_PT', 'DELAY_WAVG'],
    highlight=True,
    name="Intersections - Priority Locations",
    popup=['LABEL', 'ADT_PT', 'AM_PT', 'DELAY_WAVG'],
    marker_type="circle_marker",
    marker_kwds={
        "radius" : "6"},
    style_kwds= {
        "fillColor":"red",
        "stroke":True,
        "weight":0.6,
        "opacity":0.7})
   
links_outside.explore(
    m=map,
    legend=False,
    zoom_on_click=True,
    tooltip=['ADT_PT', 'AM_PT', 'LOS'],
    highlight=True,
    name="Links - Outside primary corridor",
    popup=['ADT_PT', 'AM_PT', 'LOS'],
    style_kwds= {
        "color":"blue",
        "weight":6,
        "opacity":0.7})

intersections_outside.explore(
    m=map,
    legend=False,
    zoom_on_click=True,
    tooltip=['LABEL', 'ADT_PT', 'AM_PT', 'DELAY_WAVG'],
    highlight=True,
    name="Intersections - Outside primary corridor",
    popup=['LABEL', 'ADT_PT', 'AM_PT', 'DELAY_WAVG'],
    marker_type="circle_marker",
    marker_kwds={
        "radius" : "6"},
    style_kwds= {
        "fillColor":"blue",
        "stroke":True,
        "weight":0.6,
        "opacity":0.7})

folium.LayerControl().add_to(map)

map
```
