Create a map showing service line freqencies- AM peak 6:00-9:00 (inbound and outbound services)

```python
# Line frequencies
condition_window = line_freq.window == '6:00-9:00'

gdf = line_freq.loc[(condition_window),:].reset_index()

gtfs.map_gdf(gdf = gdf, 
              variable = 'ntrips', 
              colors = ["#d13870", "#e895b3" ,'#55d992', '#3ab071', '#0e8955','#066a40'], 
              tooltip_var = ['route_name'] , 
              tooltip_labels = ['Route: '], 
              breaks = [5, 10, 15, 20],)
```
